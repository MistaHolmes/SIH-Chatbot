## Language translation model 

### Getting all the necessary tokens and setting the environment

```python
import os
from dotenv import load_dotenv
load_dotenv()

groq_api_key = os.getenv('GROQ_API_KEY')

## Langsmith Tracking
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"]="true"
os.environ["LANGCHAIN_PROJECT"]=os.getenv("LANGCHAIN_PROJECT")
```

### Importing necessary modules and libraries

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

### Initializing the openai/gpt-oss-120b model form groq for language translation

```python
llm_model = ChatGroq(model ="openai/gpt-oss-120b", api_key=groq_api_key)
```

### Building a Prompt Template and setting the output_parser

```python
prompt = ChatPromptTemplate.from_messages(
    [("system","Translate the following into {language}:"),
    ("user","{text}")]
)

parser = StrOutputParser()
```

### Using Langchain Expression Language (LCEL) for designing chains workflow 

```python
chain = prompt | llm_model | parser
```

### Getting the response from LLM model

```python
result = chain.invoke({'language':'hindi','text':'Good Morning, How are you?'})
print(result)
```
